```python
# read the States dataset
dataset = pd.read_csv("NIH_data_for_mining.csv", header = 0)
print(dataset.shape)

# split the dataset into features and labels
labels = dataset['State']
data = dataset.drop(['State', '# of studies'], axis = 1)
data_NIH = data[0:51]
data_NIH = data_NIH.drop('Funding', axis=1)

data_federal = data[51:102]
data_federal = data_federal.drop('Funding', axis=1)


data_industry = data[102:153]
data_industry = data_industry.drop('Funding', axis=1)


data_other = data[153:204]
data_other = data_other.drop('Funding', axis=1)


print(data_other.shape)
data_other.head(60)
```

```python
# calculate cophenetic distances for NIH-funded to find the optimal method for hierarchical clustering

proximity_matrix = pdist(data_NIH)

linkage_matrix_single = hier.linkage(data_NIH, method = 'single')
linkage_matrix_complete = hier.linkage(data_NIH, method = 'complete')
linkage_matrix_average = hier.linkage(data_NIH, method = 'average')
linkage_matrix_centroid = hier.linkage(data_NIH, method = 'centroid')
linkage_matrix_ward = hier.linkage(data_NIH, method = 'ward')

print('Single: ', hier.cophenet(linkage_matrix_single, proximity_matrix))
print('Complete', hier.cophenet(linkage_matrix_complete, proximity_matrix))
print('Average', hier.cophenet(linkage_matrix_average, proximity_matrix))
print('Centroid', hier.cophenet(linkage_matrix_centroid, proximity_matrix))
print('Ward', hier.cophenet(linkage_matrix_ward, proximity_matrix))

linkage_matrix_NIH = linkage_matrix_single
```
